# Introduction - Volunteers Oracle Precision and Recall

Notebook to calculate the oracle (Feat-BR) generated by the Volunteers precision and recall using the oracle (Feat-BR) generated by the Expert as Oracle.

# Load Libraries and Data

```python
from mod_finder_util import mod_finder_util
mod_finder_util.add_modules_origin_search_path()

import pandas as pd
import numpy as np

from modules.utils import aux_functions
from modules.utils import firefox_dataset_p2 as fd
```

# Load Datasets

```python
vol_oracle = fd.Feat_BR_Oracles.read_feat_br_volunteers_df()
exp_oracle = fd.Feat_BR_Oracles.read_feat_br_expert_df()
```

# Comparing Oracles

```python
vol_tp = len(aux_functions.get_true_positives(oracle_df=exp_oracle, output_df=vol_oracle))
vol_fn = len(aux_functions.get_false_negatives(oracle_df=exp_oracle, output_df=vol_oracle))
vol_fp = len(aux_functions.get_false_positives(oracle_df=exp_oracle, output_df=vol_oracle))

precision = vol_tp / (vol_tp + vol_fp) * 100
recall = vol_tp / (vol_tp + vol_fn) * 100

print('Volunteers Oracle Precision: {:2.4}%'.format(precision))
print('Volunteers Oracle Recall: {:2.4}%'.format(recall))
```
